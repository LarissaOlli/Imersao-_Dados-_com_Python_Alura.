```python
df['senioridade'] = df['senioridade'].replace({
    'SE': 'Senior',
    'MI': 'Médio',
    'EN': 'Júnior',
    'EX': 'Executivo'
})
df['senioridade'].value_counts()
```

```python
df['contrato'] = df['contrato'].replace({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelance'
})
df['contrato'].value_counts()
```

```python
df['tamanho_empresa'] = df['tamanho_empresa'].replace({
    'M': 'Média',
    'L': 'Grande',
    'S': 'Pequena'
})
df['tamanho_empresa'].value_counts()
```

```python
df['remoto'] = df['remoto'].replace({
    0: 'Presencial',
    100: 'Remoto',
    50: 'Híbrido'
})
df['remoto'].value_counts()
```

```python
import numpy as np

#Criação de um dataframe para usar de exemplo.
df_salarios = pd.DataFrame({
 "nome" : ['Ana', 'Bruno', 'Carlos', 'Daniele','Val'],
 "salarios" : [4000, np.nan, 5000, np.nan,100000]
})
#calcula a média salarial e substitui os nulos e arrededonda os valores.
df_salarios['salario_media'] = df_salarios['salarios'].fillna(df_salarios['salarios'].mean().round(2))

#calcula mediana e substitui os nulos pela mediana.
df_salarios['salario_mediana'] = df_salarios['salarios'].fillna(df_salarios['salarios'].median())
df_salarios
```

```python
df_temperaturas = pd.DataFrame({
  "Dia": ("Segunda", "Terça", "Quarta", "Quinta", "Sexta"),
  "Temperatura": (30, np.nan, np.nan,28,27)
})

df_temperaturas["preenchido_ffill"] = df_temperaturas["Temperatura"].ffill()
df_temperaturas
```

```python
df_temperaturas = pd.DataFrame({
  "Dia": ("Segunda", "Terça", "Quarta", "Quinta", "Sexta"),
  "Temperatura": (30, np.nan, np.nan,28,27)
})

df_temperaturas["preenchido_bfill"] = df_temperaturas["Temperatura"].bfill()
df_temperaturas
```

```python
df_cidades = pd.DataFrame({
  "nome": ['Ana', 'Bruno', 'Carlos','Daniele','Val'],
  "cidade": ("São Paulo", np.nan, "Curitiba",np.nan,"Belém")
})

df_cidades['cidade_preenchida'] = df_cidades["cidade"].fillna("Não informado")
df_cidades
```

# Aula 3 - Visualização de Dados.

```python
df_limpo.head()
```

```python
df_limpo['senioridade'].value_counts().plot(kind='bar', title="Distribuição de Senioridade")
```

```python
import seaborn as sns
```

```python
sns.barplot(data=df_limpo, x='senioridade', y='usd')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df_limpo, x='senioridade', y='usd')
plt.title("Salarios Média por Senioridade")
plt.xlabel("Senioridade")
plt.ylabel("Salario Médio Anual (USD)")
plt.show()
```

```python
df_limpo.groupby('senioridade')['usd'].mean().sort_values(ascending=False)
```

```python
ordem = df_limpo.groupby('senioridade')['usd'].mean().sort_values(ascending=True).index
```

```python
ordem
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df_limpo, x='senioridade', y='usd', order=ordem)
plt.title("Salarios Média por Senioridade")
plt.xlabel("Senioridade")
plt.ylabel("Salario Médio Anual (USD)")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df_limpo['usd'], bins=50, kde=True)
plt.title("Distribuição dos Salários Anuais")
plt.xlabel("Salário Anual (USD)")
plt.ylabel("Frequência")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x=df_limpo['usd'])
plt.title("Boxplot dos Salários Anuais")
plt.xlabel("Salário Anual (USD)")
plt.show()
```

```python
ordem_senioridade = ['Júnior', 'Médio', 'Senior', 'Executivo']
plt.figure(figsize=(10,5))
sns.boxplot(x='senioridade', y='usd', data=df_limpo, order=ordem_senioridade)
plt.title("Boxplot da Distribuição por Senioridade")
plt.xlabel("Salário Anual (USD)")
plt.show()
```

```python
ordem_senioridade = ['Júnior', 'Médio', 'Senior', 'Executivo']
plt.figure(figsize=(10,5))
sns.boxplot(x='senioridade', y='usd', data=df_limpo, order=ordem_senioridade, palette="Set2", hue="senioridade")
plt.title("Boxplot da Distribuição por Senioridade")
plt.xlabel("Salário Anual (USD)")
plt.show()
```

```python
import plotly.express as px
```

```python
df_senioridade_usd = df_limpo.groupby('senioridade')['usd'].mean().reset_index()
fig = px.bar(df_senioridade_usd, x='senioridade', y='usd', title='Salário Médio por Senioridade', labels={'senioridade': 'Senioridade', 'usd': 'Salário Médio Anual (USD)'}, category_orders={'senioridade': ordem_senioridade})
fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['remoto', 'quantidade']
fig = px.pie(remoto_contagem,
             names='remoto', # Corrigido de 'tipo_remoto' para 'remoto'
             values='quantidade',
             title='Proporção dos tipos de trabalho remoto')
fig.show()
```
